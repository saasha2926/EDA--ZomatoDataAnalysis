**IMPORTING LIBRARIES**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

**READING CSV**

```python
df = pd.read_csv(r'C:\Users\arsha amal\Downloads\zomato.csv\zomato.csv')
df.head(10)
```

```python
df.columns
```

**DROP UNWANTED COLUMNS**

```python
df = df.drop(['url','address','phone','menu_item','reviews_list','dish_liked'],errors = 'ignore')
```

```python
df.head()
```

```python
df.info()
```

**REMOVE DUPLICATES**

```python
df.drop_duplicates(inplace = True)
df
```

**HANDLE RATE ( REMOVE /5 , NEW ,NAN ,-)**

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if value == 'NEW' or value == '-':
        return np.nan
    else :
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df.head()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.head()
```

**RENAME COLUMNS APPROX COST AND LISTED IN**

```python
df.rename(columns = {'approx_cost(for two people)': 'Cost2Plates','listed_in(type)':'Type'},inplace = True)
df.head()
```

**HANDLE LISTED IN**
LOCATION AND LSITED IN HAVE SAME VALUES SO  ONE OF THEM IS DROPPED

```python
df['location'].unique
```

```python
df['listed_in(city)'].unique
```

```python
df = df.drop(['listed_in(city)'],axis = 1)
df.head()
```

**HANDLE COST2PLATES**

```python
df['Cost2Plates'].unique()
```

```python
def handlecomma(value):
    value = str(value)
    if ','in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)

df['Cost2Plates']= df['Cost2Plates'].apply(handlecomma)
df['Cost2Plates'].unique()
```

**HANDLE REST TYPE**

```python
rest_type = df['rest_type'].value_counts(ascending = False)
rest_type
```

```python
rest_type_lesser = rest_type[rest_type< 1000]
rest_type_lesser
```

```python
def handleRest_type(value):
    if value in rest_type_lesser:
        return 'others'
    else:
        return value
df['rest_type']= df['rest_type'].apply(handleRest_type)
df['rest_type'].unique()
```

**HANDLE LOCATION**

```python
location_it = df['location'].value_counts(ascending = False)
```

```python
location_lesser = location_it[location_it < 300]
location_lesser.unique()
location_lesser.value_counts().sum()
```

```python
def handle_location(value):
    if value in location_lesser:
        return 'others'
    else:
        return value

df['location'] =  df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

**HANDLE  CUISINES**

```python
 cuisine = df['cuisines'].value_counts()
cuisine
```

```python
cuisine_less = cuisine[cuisine<100]
cuisine_less
```

```python
def handleCuisine(value):
    if value in cuisine_less:
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handleCuisine)
df['cuisines'].value_counts()
```
